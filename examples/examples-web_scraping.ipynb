1) `html`에서 `사과`를 추출하여라. 


````python
from bs4 import BeautifulSoup

html = """
<html><body>
<div id = "main_goods">
  <h1>과일과 야채</h1>
  <ul id = "fr_list">
    <li class = "red">사과</li>
    <li class = "red">딸기</li>
    <li class = "purple">포도</li>
    <li class = "yellow">오렌지</li>
    <li class = "yellow">망고</li>
  </ul>
  <ul id = "ve_list">
    <li class = "red">토마토</li>
    <li class = "white">무</li>
    <li class = "black">가지</li>
    <li class = "white">연근</li>
    <li class = "yellow">당근</li>
  </ul>
</div>
</body></html>
"""

soup = BeautifulSoup(html,'html.parser')

# 사과가 li 태그의 첫번째이므로 아래 코드로 추출할 수 있다.
print(soup.li.string)
# 모든 li 태그 중 첫번째를 추출하려면 아래 코드를 사용할 수도 있다.
print(soup.find_all('li')[0].string)
# li 태그 중 class 가 red인 첫번째 것을 선택한다.
print(soup.select_one('li.red').string)
# 그냥 문자열 추출을 할 수도 있다.
print(html[100:102])
# find를 이용하여 찾을수도 있다.
ind = html.find('사과')
print(html[ind:ind+2])
````

3) `html`에서 `토마토`,`무`, `가지`, `연근`, `당근`을 추출하여 `ve_list.txt` 파일에 저장하여라. 

````python
from bs4 import BeautifulSoup

html = """
<html><body>
<div id = "main_goods">
  <h1>과일과 야채</h1>
  <ul id = "fr_list">
    <li class = "red">사과</li>
    <li class = "red">딸기</li>
    <li class = "purple">포도</li>
    <li class = "yellow">오렌지</li>
    <li class = "yellow">망고</li>
  </ul>
  <ul id = "ve_list">
    <li class = "red">토마토</li>
    <li class = "white">무</li>
    <li class = "black">가지</li>
    <li class = "white">연근</li>
    <li class = "yellow">당근</li>
  </ul>
</div>
</body></html>
"""

soup = BeautifulSoup(html,'html.parser')

# ve_list를 뽑아내는 기본형
with open('ve_list.txt','w') as f:
  for item in soup.find(id='ve_list').stripped_strings: # ve_list 아이디를 가진 ul 내부의 모든 텍스트를 가져오기, stripped_strings를 사용하여 공백 제거
    f.write(item+'\n') # 줄바꿈으로 저장

# 어려울 때 그냥 단순하게 하는 법
with open('ve_list.txt','w') as f:
  for item in soup.find_all('li')[5:]: # li 를 전부 가져온 뒤 마지막 5개에 대해서만 for문으로 저장
    f.write(item.string+'\n') # 줄바꿈으로 저장
````

4) `html`에서 `오렌지`, `망고`, `당근`을 추출하여 `yellow.txt` 파일에 저장하여라. 

````python
from bs4 import BeautifulSoup

html = """
<html><body>
<div id = "main_goods">
  <h1>과일과 야채</h1>
  <ul id = "fr_list">
    <li class = "red">사과</li>
    <li class = "red">딸기</li>
    <li class = "purple">포도</li>
    <li class = "yellow">오렌지</li>
    <li class = "yellow">망고</li>
  </ul>
  <ul id = "ve_list">
    <li class = "red">토마토</li>
    <li class = "white">무</li>
    <li class = "black">가지</li>
    <li class = "white">연근</li>
    <li class = "yellow">당근</li>
  </ul>
</div>
</body></html>
"""

soup = BeautifulSoup(html,'html.parser')

# yellow만 뽑아내는 기본형
with open('yellow.txt','w') as f:
  for item in soup.select('li.yellow'): # li 중 class가 yellow인 것들을 전부 가져옴
    f.write(item.string+'\n') # 줄바꿈으로 저장

# 조금 더 단순한 방식
with open('yellow.txt','w') as f:
  li_all = soup.find_all('li')
  for idx in [3,4,9]: # li 를 전부 가져온 뒤 마지막 5개에 대해서만 for문으로 저장
    f.write(li_all[idx].string+'\n') # 줄바꿈으로 저장
````

image.png


````python
import urllib.request
from bs4 import BeautifulSoup

url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn"

page = urllib.request.urlopen(url)

text = page.read().decode('euc-kr') # 한글이 섞여 있는 페이지의 경우 decode 방식을 'utf8' 이 아닌 'cp949' 혹은 'euc-kr'을 사용

soup = BeautifulSoup(text,'html.parser')

# 조회순으로 정렬된 영화 정보가 html 에서 어떤 특징을 가지고 있는지 우선 파악해야 한다. 위 문제의 TIP을 참고하자.
with open('movie_view.txt','w') as f:
  for title in soup.tbody.find_all('a'): # 영화 제목 정보는 tbody 태그 안에 있고, 각 영화 제목은 a 태그 안에 있음을 이용한다.
    f.write(title.string+'\n')
````

아래의 사이트를 방문하면 실시간으로 변하는 커피콩의 가격을 확인할 수 있다. 
> http://beans.itcarlow.ie/prices.html

커피콩 가격이 6.0달러 이상이면 커피숍의 아메리카노 가격을 인상하라는 문구를 출력하는 코드를 작성하여라.   
가격 확인은 1초에 한 번 하는 것으로 한다.   

시차를 두고 코드를 실행하기 위해 `time` 모듈의 `sleep()` 함수를 활용할 수 있다.   
아래와 같이 코드를 작성하면, 코드 실행을 5초 정지한다. 

````python
import time

print('안녕')
time.sleep(5)
print('하세요')
````

````python
import urllib.request
from bs4 import BeautifulSoup
import time

# 종료 시키려면 왼쪽에 정지 버튼 누르기
while True:
  try:
    url = "http://beans.itcarlow.ie/prices.html"

    page = urllib.request.urlopen(url)

    text = page.read().decode('utf8')

    soup = BeautifulSoup(text,'html.parser')

    # tip을 이용하여 커피콩의 가격이 strong 태그 안에 있음을 알 수 있다.
    price = soup.strong.string 
    price = price[1:] # 앞에 붙는 $표시를 떼어준다.
    price = float(price) # 실수로 변환

    print("현재 커피콩 가격은",price,"달러 입니다.") # 없어도 되지만 확인차
    if price >= 6.0:
      print("아메리카노 가격을 인상하세요")
    else:
      print("아메리카노 가격을 인상하지 않으셔도 됩니다.") # 없어도 되지만 확인차 작성

    time.sleep(1) # 1초 간격으로 확인

  except: # 페이지가 새로고침 되는 과정에서 가끔 None이 넘어오는 경우가 있는데, 이를 방지하기 위해서 try except문 사용
    continue
````

**[추가문제2]**
아래는 코로나바이러스감염증-19 정식 홈페이지이다. 

> http://ncov.mohw.go.kr/

일일 국내 발생 확진자 수를 추출하여라. 

image.png


````python
import urllib.request
from bs4 import BeautifulSoup

url = "http://ncov.mohw.go.kr/"

page = urllib.request.urlopen(url)

text = page.read().decode('utf8') # 이 페이지는 한글이 섞여있지만 cp949나 euc-kr로는 디코딩이 안됨. utf8 디코딩이 가능하므로 utf8로 디코딩

soup = BeautifulSoup(text, 'html.parser')

# tip을 이용하여 일일 확진자 정보가 div 태그,class 는 liveNum_today_new 안에 있음을 알 수 있다.
liveNum_today_new = soup.select_one('div.liveNum_today_new')
# 그 중 국내 발생 명수는 span 태그, class는 data 안에 있는 첫번째 정보임을 알 수 있다. (두번째는 해외 유입 명수)
Domestic = liveNum_today_new.select('span.data')[0].string
print("일일 국내 발생 확진자 수는",Domestic,'명 입니다.')
````
